```python
import librosa
import matplotlib.pyplot as plt
```

```python
# Daftar file audio
audio_files = [
    "faded_audio_1.mp3",
    "faded_audio_2.mp3",
    "audio_fast.wav",
    "audio_slow.wav",
    "podcast_with_fade.mp3",
    "fast_fade_audio.wav"
]

# Membuat subplots untuk 6 file audio
plt.figure(figsize=(12, 10))

for i, audio_file in enumerate(audio_files, 1):
    # Load audio
    y, sr = librosa.load(audio_file)
    
    # Buat subplot untuk setiap file audio
    plt.subplot(3, 2, i)
    librosa.display.waveshow(y, sr=sr)
    plt.title(f"Waveform {audio_file}")
    plt.xlabel("Waktu (detik)")
    plt.ylabel("Amplitude")

# Tampilkan semua waveform
plt.tight_layout()
plt.show()
```
